```python
import plotly.graph_objects as go
```

```python
def plot_innings_total(df):
    fig = go.Figure()

    # Add first innings trace
    fig.add_trace(
        go.Scatter(
            x=df["date"],
            y=df["innings1_score"],
            mode="lines",
            name="First Innings Score",
            line=dict(color="blue"),
        )
    )

    fig.add_trace(
        go.Scatter(
            x=df["date"],
            y=df["innings2_score"],
            mode="lines",
            name="First Innings Score",
            line=dict(color="red"),
        )
    )

    fig.update_layout(
        title="First And Second Innings Total and Wickets In IPL 2024",
        xaxis_title="Date",
        yaxis_title="Score",
        plot_bgcolor="white",
        paper_bgcolor="white",
        font=dict(color="black"),
        height=600,
        width=1000,
    )

    fig.show()


plot_innings_total(df)
```

```python
def plot_innings_histogram(df,year=None):
    plt.figure(figsize=(18, 9),dpi=300)  # Set figure size (width, height)

    # Plot histogram for first innings scores
    plt.hist(
        df["innings1_score"],
        bins=5,  
        color="#1f77b4",
        alpha=0.65, 
        label="First Innings Scores",
        edgecolor="black"
    )

    # Customize the plot
    plt.title(f"Distribution of First Innings Scores in IPL {year}", pad=20,color="black")
    plt.xlabel("Runs Scored")
    plt.ylabel("Number of Matches")
    plt.grid(True, linestyle="--", linewidth=0.5, color="lightgrey")
    plt.legend(loc="upper right",labelcolor="black",frameon=True,facecolor="white")

    # Save the plot as PNG
    plt.savefig("innings_distribution.png", dpi=200, bbox_inches="tight")
    plt.show()  # Display the plot


plot_innings_histogram(df)
```

```python
def get_frist_innings_total(df):
    innings_total = df.groupby(by=["date","match_id", "innings"], as_index=False).sum(numeric_only=True)
    innings_total["total_runs"] = (
        innings_total["runs_of_bat"]
        + innings_total["extras"]
    )
    innings_total_1 = innings_total[innings_total["innings"] == 1]
    innings_total_1 = innings_total_1.rename(columns={"total_runs": f"innings1_score"})
    return innings_total_1
```

```python
plot_innings_histogram(get_frist_innings_total(load_csv("data/cricketData/ipl_2022_deliveries.csv")),2022)
plot_innings_histogram(get_frist_innings_total(load_csv("data/cricketData/ipl_2023_deliveries.csv")), 2023)
plot_innings_histogram(get_frist_innings_total(load_csv("data/cricketData/ipl_2024_deliveries.csv")), 2024)
```
